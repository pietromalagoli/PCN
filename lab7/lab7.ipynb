```r
library(ggplot2)
library(RColorBrewer)
library(rgl)
options(repr.plot.width=15, repr.plot.height=10)

source("/home/pietromalagoli/PCN/common.R")
mypal <- viridis::viridis_pal()(20)
set.seed(12345)

g <- sample_smallworld(dim=2, size = 20, p = 0.05, nei=1)
lay2D <- layout_with_fr(g, dim=2)
lay3D <- layout_with_fr(g, dim=3)
deg <- degree(g)
centr_deg <- deg
cols_deg <- vec2pal(centr_deg, mypal)
clear3d()
rglplot(g, layout = lay3D, vertex.color = cols_deg, vertex.size = deg, vertex.label = NA, edge.color = "gray80", vertex.frame.color = NA, edge.width = 0.5)
#plot(g, layout=lay3D, vertex.color=cols_deg, vertex.size=deg, vertex.label=NA, edge.color="gray80", vertex.frame.color=NA, edge.width=0.5)
```

```r
centr_betw <- betweenness(g, normalized = TRUE)
cols_betw <- vec2pal( log(1 + centr_betw), mypal)
clear3d()
plot(g, layout=lay3D, vertex.color=cols_deg, vertex.size=deg, vertex.label=NA, edge.color="gray80", vertex.frame.color=NA, edge.width=0.5)
```

```r
centr_close <- closeness(g, normalized = TRUE)

centr_pr <- page_rank(g)$vector
centr_eig <- eigen_centrality(g)$vector
centr_hub <- hub_score(g)$vector
centr_auth <- authority_score(g)$vector

df <- data.frame(
    Degree = centr_deg,
    Betweenness = centr_betw,
    Closeness = centr_close,
    PageRank = centr_pr,
    Eigenvector = centr_eig,
    Hub = centr_hub,
    Authority = centr_auth
)
df
```

# Quantitative Comparison #

```r
N <- 1000
OM = matrix(0.001, ncol=4, nrow=4)
diag(OM) = 0.03
g = sample_sbm(N, pref.matrix = OM, block.sizes = c(200, 200, 300, 300))
```

```r
plot(g)
```

```r
lay = layout_with_fr(g)
plot(g, layout=lay, vertex.size=degree(g))
```

```r
planted_partition = c(rep(1,200), rep(2,200), rep(3,300), rep(4, 300))
comms_lou = cluster_louvain(g, resolution=1)
comms_spin = cluster_spinglass(g, gamma=1)
comms_info = cluster_infomap(g)


par(mfrow=c(1,3), mar=c(0,0,1,1))
plot(g, layout=lay, vertex.color=comms_lou$membership, vertex.size=degree(g), vertex.label=NA, edge.color="gray80", vertex.frame.color="white", edge.width=0.3, main=paste("Louvain - ", round(modularity(comms_lou), 4)))
plot(g, layout=lay, vertex.color=comms_spin$membership, vertex.size=degree(g), vertex.label=NA, edge.color="gray80", vertex.frame.color="white", edge.width=0.3, main=paste("Spin - ", round(modularity(comms_spin), 4)))
plot(g, layout=lay, vertex.color=comms_info$membership, vertex.size=degree(g), vertex.label=NA, edge.color="gray80", vertex.frame.color="white", edge.width=0.3, main=paste("Infomap - ", round(modularity(comms_info), 4)))
```

```r
compare(planted_partition, comms_lou$membership, method="nmi") # nmi: normalized mutual information
compare(planted_partition, comms_spin$membership, method="nmi")
compare(planted_partition, comms_info$membership, method="nmi")
```

```r
df_mod = data.frame()
for (gam in 10^seq(-2,1,0.25)) {
    comms_lou = cluster_louvain(g, resolution=gam)
    modularity = modularity(comms_lou)
    groups = length(unique(membership(comms_lou)))

    cat(paste("Groups:", groups, "\n"))
    df_mod = rbind(df_mod, data.frame(gamma=gam, modularity=modularity, groups = groups))
}
ggplot(df_mod, aes(x = gamma, y = modularity)) +
    theme_bw() +
    geom_line()
```
